```python
# split up span into start and end date + career length
df['Span'].str.split(pat ='-')
```

```python
df['Start_year']=df['Span'].str.split(pat ='-').str[0]
```

```python
df['End_year']=df['Span'].str.split(pat ='-').str[1]
```

```python
# drop span columns

df= df.drop(['Span'], axis = 1)
```

```python
# Split the country name  from Player

df['Player'].str.split(pat ='(')
```

```python
# We first split the strings in the 'Player' column based on the '('  
#  we extract the second element of the resulting lists using str[1]  similarly for ohter side but wit .str[0]


df['Country']=df['Player'].str.split(pat ='(').str[1].str.split(')').str[0]
```

```python

df['Player']=df['Player'].str.split(pat ='(').str[0]
```

```python
# Highest_inns_Score
# Balls_Faced	
# Batting _Strike_Rate
# Start_year	
# End_year


df['Highest_inns_Score']=df['Highest_inns_Score'].str.split(pat ='*').str[0]
```

```python


df['Highest_inns_Score']=df['Highest_inns_Score'].astype('int')
```
